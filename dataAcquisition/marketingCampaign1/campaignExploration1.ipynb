```python
import pandas as pd

db = pd.read_csv('/Users/santi/Library/CloudStorage/OneDrive-UniversidadPontificiaComillas/Uni/4/TFG/Proyecto/dataAcquisition/marketingCampaign1/campaign1.csv')
db.head()
```

```python
db.columns
```

```python
db.drop(['campaign_item_id', 'ext_service_id', 'creative_id',
       'advertiser_id',
       'channel_id', 'channel_name', 'timezone', 'search_tags', 'template_id'], axis=1, inplace=True)
```

```python
db.head()
```

```python
db['time'] = pd.to_datetime(db['time'])

db['year'] = db['time'].dt.year
db['month'] = db['time'].dt.month
db['day'] = db['time'].dt.day

db.drop(['time'], axis=1, inplace=True)

db.head()
```

```python
db.columns
```

```python
db_dummies = pd.get_dummies(db, columns=['weekday_cat', 'ext_service_name'], drop_first=True)
db_dummies.columns
```

```python
db_dummies.rename(columns={'weekday_cat_week_day': 'weekDay', 'weekday_cat_week_end': 'weekeEnd'}, inplace=True)
db_dummies.columns
```

```python
#I could keep going treating the data but I think it is enough for an exploration

import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = db_dummies.select_dtypes(include=['number', 'bool']).corr()
plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
