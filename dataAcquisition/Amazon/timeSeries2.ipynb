```python
import pandas as pd
df = pd.read_excel('./AmazonSaleReport.xlsx', parse_dates=['Date'], index_col='Date')
df.sort_index(inplace=True)
df.head()
```

```python
df_timeseries = df['Amount']

# transform date into datetime object
df_timeseries['Date'] = pd.to_datetime(df['Date'])
df_timeseries.head()
```

```python
#sum the amount for each day
df_time_series = df.resample('D').sum()['Amount']

df_time_series.head()
```
